```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
x = df[['Age', 'EstimatedSalary']]
```

```python
x
```

```python
plt.figure(figsize=(16,9))

plt.title('data distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Age')
plt.scatter(x['EstimatedSalary'], x['Age'], color =  'red');
plt.grid()
```

```python
x.describe()
```

#### normalization

```python
# import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# crete an object
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled;
```

```python
x_scaled[:,0].min(), x_scaled[:,0].max()
```

```python
x_scaled[:,1].min(), x_scaled[:,1].max()
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
plt.figure(figsize=(16,9))

plt.title('data distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Age')
plt.scatter(x_scaled[:,1], x_scaled[:,0], color =  'red');
plt.grid()
```
