```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([12, 17, 23, 29, 33, 39, 44, 51, 59, 64])
y = np.array([14, 19, 24, 30, 37, 41, 47, 55, 59, 61])
z = np.array([64, 59, 51, 49, 42, 36, 32, 28, 22, 10])
```

#### line plot

```python
plt.title('change in y w.r.t x')
plt.xlabel('x values')
plt.ylabel('y values')
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--');
plt.grid();
```

```python
plt.title('change in z w.r.t x')
plt.xlabel('x values')
plt.ylabel('z values')
plt.plot(x, z, marker = 'o', mfc = 'red', ls = '--');
plt.grid();
```

```python
plt.title('line plot')
plt.xlabel('x values')
plt.ylabel('y-z values')
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--', color = 'black',
        label = "2023");
plt.plot(x, z, marker = 'o', mfc = 'black', ls = '--', color = 'purple',
        label = '2024');
plt.legend();ot(y)           # plot y using x as index array 0..
plt.grid();
```

```python
plt.figure(figsize= (16,9))

plt.subplot(2,2,1)
plt.title('change in y w.r.t x')
plt.xlabel('x values')
plt.ylabel('y values')
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--');
plt.grid();

plt.subplot(2,2,2)
plt.title('change in z w.r.t x')
plt.xlabel('x values')
plt.ylabel('z values')
plt.plot(x, z, marker = 'o', mfc = 'red', ls = '--');
plt.grid();
```

#### Box plot

```python
plt.boxplot(df['Age']);
```

```python
q3 = np.percentile(df['Age'], 75)
q1 = np.percentile(df['Age'], 25)
q3, q1
```

```python
iqr = q3 - q1 
iqr
```

```python
ub = q3 + 1.5 * iqr
lb = q1 - 1.5 * iqr
ub, lb
```

```python
df = df[(df['Age'] > lb) & (df['Age'] < ub)]
```
